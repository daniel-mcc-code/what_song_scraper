```python
import pandas as pd

soundtracks = pd.read_json("/Users/danielmccarthy/coding_projects/week_3/what_song_scraper/what_song/what_song/what_song_search.jsonl", lines=True)

songs_df = pd.json_normalize(soundtracks['props'], sep='_').explode("pageProps_movie_songs")

songs_df = soundtracks.merge(songs_df, left_index=True, right_index=True)

songs_df

# songs_df.to_csv("songs.csv", index=False)

```

```python
songs_normalised = pd.json_normalize(songs_df["pageProps_movie_songs"])

songs_normalised

# songs_normalised.to_csv("songs_normalised.csv", index=False)
```

```python
# Drop irrelevant and empty columns
columns_to_drop = [
    "props", "buildId", "isFallback", "isExperimentalCompile", "gssp", "scriptLoader",
    "redirect_times", "redirect_ttl", "redirect_urls", "redirect_reasons",
    "artist_name", "highlighted_artist_name"  # Completely empty columns
]
 # Copy of soundtracks dataframe to movies
movies = soundtracks

movies_cleaned = movies.drop(columns=columns_to_drop)

# Rename columns for clarity
movies_cleaned = movies_cleaned.rename(columns={
    "_score": "relevance_score"
})

# Ensure 'year_released' column exists before converting its data type
if 'year' in movies_cleaned.columns:
    movies_cleaned["year"] = movies_cleaned["year"].astype(int)

movies_cleaned.to_csv("movies_cleaned.csv", index=False)
```

```python
# Merge movies_cleaned and songs_normalised on 'id'
movies_songs = movies_cleaned.merge(songs_normalised, left_on="id", right_on="movie_id")

movies_songs.to_csv("movies_songs.csv", index=False)
```

```python
# Drop irrelevant columns from movies_songs
columns_to_drop = [
    "page", "query", "img", "highlighted_title", "year", "download_timeout", "download_slot", "download_latency", "depth", "movie_id", "played_count", "user_scene_id", "user_time_id", "user_song_id", "ts_id", "ts_link", "ts_thumbnail", "song.previewUrl", "song.spotifyImg64", "song.appleMusicPreviewUrl", "song.youtube_id", "song.track_number", "song.artist.__typename", "song.__typename", "__typename"
]
movies_songs_cleaned = movies_songs.drop(columns=columns_to_drop)

# Rename columns for clarity

movies_songs_cleaned = movies_songs_cleaned.rename(columns={
    "title": "movie_title",
    "year": "movie_release_year",
    "id_x": "whatsong_movie_id",
    "relevance_score": "whatsong_movie_relevance_score",
    "entry_id": "whatsong_entry_id",
    "id_y": "entry_id",
    "time_description": "mins_into_movie",
    "is_trailer": "in_trailer",
    "song.id": "whatsong_song_id",
    "song.artist.id": "whatsong_artist_id",
    "song.artist.name": "artist_name",
    "song.name": "song_title",
    "song.artist": "song_artist",
    "song.artist.slug": "song_artist_slug",
    "song.spotifyUri": "song_spotify_uri",
    "song.appleMusicUrl": "song_apple_music_url",
    "song.contributor_user_id": "whatsong_contributor_user_id",
})

display(movies_songs_cleaned)

movies_songs_cleaned.to_csv("movies_songs_cleaned.csv", index=False)
```
