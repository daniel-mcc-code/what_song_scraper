```python
import pandas as pd

soundtracks = pd.read_json("/Users/danielmccarthy/coding_projects/week_3_movies/what_song_scraper/what_song/what_song/what_song_search.jsonl", lines=True)

songs_df = pd.json_normalize(soundtracks['props'], sep='_').explode("pageProps_movie_songs")

songs_df = soundtracks.merge(songs_df, left_index=True, right_index=True)

songs_df

# songs_df.to_csv("songs.csv", index=False)

```

```python
songs_normalised = pd.json_normalize(songs_df["pageProps_movie_songs"])

songs_normalised

# songs_normalised.to_csv("songs_normalised.csv", index=False)
```

```python
# Drop irrelevant and empty columns
columns_to_drop = [
    "props", "buildId", "isFallback", "isExperimentalCompile", "gssp", "scriptLoader",
    "redirect_times", "redirect_ttl", "redirect_urls", "redirect_reasons",
    "artist_name", "highlighted_artist_name"  # Completely empty columns
]
 # Copy of soundtracks dataframe to movies
movies = soundtracks

movies_cleaned = movies.drop(columns=columns_to_drop)

# Rename columns for clarity
movies_cleaned = movies_cleaned.rename(columns={
    "_score": "relevance_score"
})

# Ensure 'year_released' column exists before converting its data type
if 'year' in movies_cleaned.columns:
    movies_cleaned["year"] = movies_cleaned["year"].astype(int)

movies_cleaned.to_csv("movies_cleaned.csv", index=False)
```

```python
# Merge movies_cleaned and songs_normalised on 'id'
movies_songs = movies_cleaned.merge(songs_normalised, left_on="id", right_on="movie_id")

# movies_songs.to_csv("movies_songs.csv", index=False)
```

```python
# Drop irrelevant columns from movies_songs
columns_to_drop = [
    "type", "page", "query", "img", "highlighted_title", "year", "download_timeout", "download_slot", "download_latency", "depth", "movie_id", "played_count", "user_scene_id", "user_time_id", "user_song_id", "ts_id", "ts_link", "ts_thumbnail", "song.previewUrl", "song.spotifyImg64", "song.appleMusicPreviewUrl", "song.youtube_id", "song.track_number", "song.artist.__typename", "song.__typename", "__typename"
]
movies_songs_cleaned = movies_songs.drop(columns=columns_to_drop)

# Rename columns for clarity

movies_songs_cleaned = movies_songs_cleaned.rename(columns={
    "title": "movie_title",
    "year": "movie_release_year",
    "id_x": "whatsong_movie_id",
    "relevance_score": "whatsong_movie_relevance_score",
    "entry_id": "whatsong_entry_id",
    "id_y": "entry_id",
    "time_description": "mins_into_movie",
    "is_trailer": "in_trailer",
    "song.id": "whatsong_song_id",
    "song.artist.id": "whatsong_artist_id",
    "song.artist.name": "artist_name",
    "song.name": "song_title",
    "song.artist": "song_artist",
    "song.artist.slug": "song_artist_slug",
    "song.spotifyUri": "song_spotify_uri",
    "song.appleMusicUrl": "song_apple_music_url",
    "song.contributor_user_id": "whatsong_contributor_user_id",
})

display(movies_songs_cleaned)

# movies_songs_cleaned.to_csv("movies_songs_cleaned.csv", index=False)
```

```python
# Merge movies_songs_cleaned with box_office on 'movie_title' and 'title'

# Load box_office data
box_office = pd.read_csv("box_office.csv")

# Merge movies_songs_cleaned with box_office on 'movie_title'
movies_songs_box_office = movies_songs_cleaned.merge(box_office, left_on="movie_title", right_on="title", suffixes=("", "_y"))


# Merge movies_songs_box_office with songs_df[pageProps_releaseDate] on 'movie_title' and "pageProps_movie_name" to get movie release date
movies_songs_box_office = movies_songs_box_office.merge(songs_df[["pageProps_movie_releaseDate", "pageProps_movie_name"]], left_on="movie_title", right_on="pageProps_movie_name", how="left")

# Drop duplicates again after the second merge
movies_songs_box_office = movies_songs_box_office.drop_duplicates()

# Drop irrelevant columns and rename pageProps_movie_releaseDate and release_year_y to movie_release_date
movies_songs_box_office = movies_songs_box_office.drop(columns=[
    "pageProps_movie_name", "slug", "song_artist_slug", "title", "url", "link", "release_year_y", "rank"])

movies_songs_box_office = movies_songs_box_office.rename(columns={"pageProps_movie_releaseDate": "movie_release_date"})

# Convert 'movie_release_date' to datetime but just days, months and years
movies_songs_box_office["movie_release_date"] = pd.to_datetime(movies_songs_box_office["movie_release_date"], errors="coerce").dt.strftime("%Y-%m-%d")

# Reset index
movies_songs_box_office = movies_songs_box_office.reset_index(drop=True)

display(movies_songs_box_office)
```

```python
# Rearrange columns
movies_songs_box_office = movies_songs_box_office[[
    "movie_title", "release_year", "movie_release_date", "mins_into_movie", "artist_name", "song_title", "scene_description", "in_trailer", "worldwide_gross", "domestic_gross", "foreign_gross", "whatsong_movie_relevance_score", "song_spotify_uri", "song_apple_music_url", "whatsong_contributor_user_id", "whatsong_movie_id", "whatsong_artist_id", "whatsong_song_id", "entry_id",
]]


# export to csv
movies_songs_box_office.to_csv("movies_songs_box_office.csv", index=False)

display(movies_songs_box_office)

```
