```python
import pandas as pd

# Load in json data
what_song = pd.read_json("/Users/danielmccarthy/coding_projects/week_3_movies/what_song_scraper/what_song/what_song/what_song_tv.json")

episodes_df = pd.json_normalize(what_song['pageProps'], sep='_')

# Export to csv
episodes_df.to_csv('episodes.csv', index=False)

# Drop irrelevant columns
columns_to_drop = ["statsigProps", "key", "_sentryTraceData", "_sentryBaggage", "episode_tvshow_similarities", "episode_tmdb_banner_url", "episode_tvshow_bannerUrlTmdb", "episode_tvshow_posterUrlTmdb", "episode_tvshow_contributor_user", "episode_nextEpisode", "episode_prevEpisode___typename", "episode_discussions",
       "episode_nextEpisode___typename", "episode_nextEpisode_id",
       "episode_nextEpisode_episode_number" ]

episodes_df = episodes_df.drop(columns=columns_to_drop)


display(episodes_df)
```

```python
# Explode the songs column
songs_df = episodes_df.explode('episode_songs')

# Drop episode songs column
episodes_df = episodes_df.drop(columns=['episode_songs'])

# Normalize the songs column
songs_df = pd.json_normalize(songs_df['episode_songs'], sep='_')

songs_df.to_csv('songs.csv', index=False)

songs_df
```

```python

# Merge the two dataframes
tv_songs_df = pd.merge(episodes_df, songs_df, on="episode_id")

# Drop irrelevant columns if they exist
columns_to_drop2 = ['episode___typename', 'episode_tvshow___typename', 'episode_tvshow_slug', 'episode_tvshow_music_supervisor_to_title', 'episode_prevEpisode_id', 'episode_prevEpisode_episode_number', 'episode_season___typename', 'episode_tvshow_theme_song_youtube_id', 'episode_prevEpisode', '__typename', 'song___typename', 'user_scene_id', 'user_time_id', 'user_song_id', 'ts_id', 'ts_link', 'ts_thumbnail', 'song_previewUrl', 'song_artist_slug', 'song_spotifyImg300', 'song_spotifyImg64', 'song_appleMusicPreviewUrl', 'song_youtube_id', 'song_about' 'song_releaseYear','episode_tvshow_theme_song_artist_slug','episode_tvshow_theme_song_slug', 'episode_tvshow_theme_song_previewUrl','episode_tvshow_theme_song_spotifyImg64', 'episode_tvshow_theme_song_appleMusicPreviewUrl', 'played_count', 'id'] # was not sure what 'id' referred to so I dropped it
columns_to_drop2 = [col for col in columns_to_drop2 if col in tv_songs_df.columns]

tv_songs_df = tv_songs_df.drop(columns=columns_to_drop2)

# Save to csv
tv_songs_df.to_csv('tv.csv', index=False)

display(tv_songs_df.columns)
tv_songs_df


```

```python
# Rename columns
tv_songs_df = tv_songs_df.rename(columns={'episode_season_id': 'whatsong_season_id', 
                                          
                                          'episode_tvshow_id': 'whatsong_tvshow_id', 
                                          'episode_episode_number': 'episode_number',
                                          'episode_air_date': 'episode_air_date',
                                          'episode_tmdb_episode_id': 'tmdb_episode_id', 
                                          'episode_tvshow_name': 'tvshow_name', 
                                          'episode_tvshow_year': 'tvshow_year_of_first_season',
                                          'episode_tvshow_composer': 'tvshow_composer', 
                                          'episode_tvshow_firstAirDate': 'tvshow_first_air_date',
                                          'episode_tvshow_network': 'tvshow_network', 
                                          'episode_tvshow_song_count': 'tvshow_song_count',
                                          'episode_tvshow_tmdbId': 'tvshow_tmdbId', 
                                          'episode_tvshow_imdbId': 'tvshow_imdbId', 
                                          'episode_tvshow_status': 'tvshow_status',
                                          'episode_tvshow_view_count': 'tvshow_view_count',
                                          'episode_tvshow_theme_song_id': 'whatsong_tvshow_theme_song_id', 'episode_tvshow_theme_song': 'tvshow_theme_song', 'episode_tvshow_contributor_user_id': 'whatsong_tvshow_contributor_user_id', 'episode_season_season_number': 'season_number', 'episode_tvshow_theme_song_name':'tvshow_theme_song_name', 'episode_tvshow_theme_song_artist_id': 'whatsong_tvshow_theme_song_artist_id', 'episode_tvshow_theme_song_artist_name': 'tvshow_theme_song_artist_name', 'song_id': 'whatsong_song_id',
                                          'scene_description': 'scene_description', 
                                          'time_description': 'mins_into_episode', 
                                          'song_artist_id': 'whatsong_song_artist_id', 
                                          'song_spotifyUri': 'song_spotify_uri', 'song_appleMusicUrl': 'song_apple_music_url', 'song_releaseYear': 'song_release_year', 'song_contributor_user_id': 'whatsong_song_contributor_user_id',
                                          'epiosde_id': 'whatsong_episode_id',
                                          
})

# Convert episode air date and tvshow_first_air_date  to datetime
tv_songs_df['episode_air_date'] = pd.to_datetime(tv_songs_df['episode_air_date']).dt.strftime("%Y-%m-%d")
tv_songs_df['tvshow_first_air_date'] = pd.to_datetime(tv_songs_df['tvshow_first_air_date']).dt.strftime("%Y-%m-%d")


# Reorder columns
new_column_order = [
    # General episode and show info
    "tvshow_name", "tvshow_year_of_first_season", "tvshow_first_air_date", "episode_name", "episode_number", 
    "episode_air_date", "season_number", "tvshow_network", "mins_into_episode", 

    # Song and artist details
    "song_name", "song_artist_name", "scene_description", "tvshow_status", "tvshow_song_count", "whatsong_song_id", "whatsong_song_artist_id", "whatsong_season_id", "whatsong_tvshow_id", "whatsong_episode_id", "tmdb_episode_id","whatsong_song_contributor_user_id", "song_spotify_uri", "song_apple_music_url", "song_about", "song_release_year", 

    # Show metadata and relevance
    "tvshow_tmdbId", "tvshow_imdbId", "tvshow_view_count",

    # Theme song details (moved to the far right)
    "tvshow_theme_song_name", "tvshow_theme_song_artist_name", "whatsong_tvshow_theme_song_artist_id", 
    "tvshow_theme_song", "whatsong_tvshow_theme_song_id", "whatsong_tvshow_contributor_user_id", 
    "episode_tvshow_theme_song_appleMusicUrl",
]

new_column_order = [col for col in new_column_order if col in tv_songs_df.columns]

# Reorder DataFrame columns
tv_songs_df = tv_songs_df[new_column_order]

tv_songs_df = tv_songs_df.sort_values(by=["tvshow_name", "season_number", "episode_number"])


# Save to csv
tv_songs_df.to_csv('tv2.csv', index=False)

tv_songs_df

```
